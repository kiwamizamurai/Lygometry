```python
xdata = np.array([-5,-2,-1,1,3,5])
ydata = np.array([-9,-2,-0.5,5,4,11])
plt.scatter(xdata,ydata)
plt.xlabel("xdata")
plt.ylabel("ydata")
plt.annotate("(x_1,y_1)",
            xy=(xdata[0], ydata[0]),xytext=(-5, -13),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
plt.annotate("(x_2,y_2)",
            xy=(xdata[1], ydata[1]),xytext=(-3, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
plt.annotate("(x_3,y_3)",
            xy=(xdata[2], ydata[2]),xytext=(-1, -9),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
plt.annotate("(x_4,y_4)",
            xy=(xdata[3], ydata[3]),xytext=(0, 10),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
plt.annotate("(x_5,y_5)",
            xy=(xdata[4], ydata[4]),xytext=(3, -5),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
plt.annotate("(x_6,y_6)",
            xy=(xdata[5], ydata[5]),xytext=(4, 13),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3"),
            )
```

```python
a = xdata
b = np.ones(len(xdata)).T
V = np.c_[b, a]
```

```python
alpha, beta = np.linalg.lstsq(V,ydata)[0]
print alpha, beta 
def fitfunc(x):
    return alpha*x + beta
```
